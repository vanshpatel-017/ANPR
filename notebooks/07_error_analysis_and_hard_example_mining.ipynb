# 07 — Error Analysis and Hard-Example Mining

Build review queues from confidence slices and prepare OCR hard-example training files.

```python
import shutil
from pathlib import Path

import pandas as pd
```

```python
INPUT_CSV = Path('unlabeled_predictions.csv')
VAL_IMAGES_DIR = Path('C:/path/to/private/data/DATASET/val/images')
REVIEW_DIR = Path('ocr_review')
REVIEW_CSV = Path('ocr_review.csv')

REVIEW_DIR.mkdir(parents=True, exist_ok=True)
df = pd.read_csv(INPUT_CSV)
slice_df = df[(df['ocr_confidence'] >= 0.90) & (df['ocr_confidence'] < 0.98)].copy()

rows = []
for _, row in slice_df.iterrows():
    image_name = row['image_name']
    src = VAL_IMAGES_DIR / image_name
    dst = REVIEW_DIR / image_name
    if src.exists():
        shutil.copy2(src, dst)
        rows.append({'image_name': image_name, 'predicted_plate': row['predicted_plate'], 'ocr_confidence': row['ocr_confidence'], 'correct_plate': ''})

pd.DataFrame(rows).to_csv(REVIEW_CSV, index=False)
print('saved =', REVIEW_CSV)
print('review_count =', len(rows))
```
